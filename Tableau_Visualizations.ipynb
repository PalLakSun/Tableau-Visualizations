9.Explain why different charts were used for

Ans-

1.Different charts were used because each business question requires a specific

type of analysis, and the right visualization helps in better understanding and

decision-making.

2.Different charts were selected based on the nature of the analysis. Bar

charts were used for comparison, line charts for trends over time, and scatter

plots for identifying relationships between variables. This ensures that each

business question is answered in the most effective and visually clear way.

3.No single chart can answer all questions

4.Each chart is chosen based on:

-Type of data

-Business question

-Insight required

-Best for comparing values across categories

Bar charts make it easy to:

Compare heights

Identify highest and lowest values quickly

2. Line Chart - Sales Trend Over Time

Best for time-based analysis

Ex-How do sales change over time?

It Shows:

Trends

Seasonality

Growth/decline patterns

3.Scatter Plot - Sales vs Profit

Best for relationship analysis between two measures

Ex-Are high sales generating high profit?

Helps identify:

Loss-making orders

Low-profit areas

Correlation between variables

To enable interactive analysis

Drill down into specific data
